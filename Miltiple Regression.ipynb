```python
#ds[["X1_Experience","X2_TestScore"]]
x = ds.iloc[:,:-1]
y = ds["Y_Salary"]
```

```python
import sklearn
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=82)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train,y_train)
```

```python
lr.predict([[93 	,90]])
```

```python
lr.score(x_test,y_test)*100
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
y_pred = lr.predict(x)
y_pred
```

```python
sns.scatterplot(x="X1_Experience",y="Y_Salary",data=ds,color="red")
plt.plot(ds["X1_Experience"],y_pred,color="cyan")
sns.scatterplot(x="X2_TestScore",y="Y_Salary",data=ds,color="red")
plt.plot(ds["X2_TestScore"],y_pred,color="cyan")
plt.legend(["org_data","predict_line"])
plt.savefig("predict.jpg")
plt.show()
```
