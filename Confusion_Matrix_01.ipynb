## Confusion matrix



    A confusion matrix is a simple and useful tool for understanding the performance of a classification models,like one used in machine learning or statistics.

    It helps you evaluate how well your models is doing in categorizing thins correctly.

    It also known as the error matrix.

    The matrix consists of predictions result in a summarized form, which has the number of correct predictions and incorrect predictions.





```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
ds = pd.read_excel(r"C:\Users\student\Downloads\SMLP Practicals\Excel data\placement_logistic_data_01.xlsx")
```

```python
ds.head()
```

```python
x= ds.iloc[:,:-1]
y= ds["Placed"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
lr.fit(x_train,y_train)
```

```python
lr.score(x_train,y_train)*100,lr.score(x_test,y_test)*100
```

```python
lr.predict(x)
```

```python
from sklearn.metrics import confusion_matrix,precision_score,recall_score,f1_score
```

```python
cf = confusion_matrix(y_test,lr.predict(x_test))
```

```python
cf
```

```python
sns.heatmap(cf,annot=True)
```

```python
precision_score(y_test,lr.predict(x_test))*100
```

```python
recall_score(y_test,lr.predict(x_test))*100
```

```python
f1_score(y_test,lr.predict(x_test))*100
```
