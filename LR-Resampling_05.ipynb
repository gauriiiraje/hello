## LR-Resampling

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from mlxtend.plotting import plot_decision_regions
```

```python
ds = pd.read_excel(r"C:\Users\student\Downloads\SMLP Practicals\Excel data\placement_logistic_data.xlsx")
```

```python
ds.head(10)
```

```python
ds["Placed"].value_counts()
```

```python
x=ds.iloc[:,:-1]
y = ds["Placed"]
```

## Random Sampling

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)*100
```

```python
lr.predict([[9.75, 67.53]])
```

## Random Under Sampling

```python
from imblearn.under_sampling import RandomUnderSampler
```

```python
ru =  RandomUnderSampler()
ru_x, ru_y = ru.fit_resample(x,y)
```

```python
ru_x,ru_y
```

```python
ru_y.value_counts()
```

```python
x_train,x_test,y_train,y_test = train_test_split(ru_x,ru_y,test_size=0.2,random_state=42)
```

```python
lr1 = LogisticRegression()
lr1.fit(x_train,y_train)
```

```python
lr1.score(x_test,y_test)*100
```

```python
lr1.predict([[9.75, 67.53]])
```

## Random Over Sampler

```python
from imblearn.over_sampling import RandomOverSampler
```

```python
ro = RandomOverSampler()
ro_x, ro_y = ro.fit_resample(x,y)
```

```python
ro_x,ro_y
```

```python
ro_y.value_counts()
```

```python
x_train,x_test,y_train,y_test = train_test_split(ro_x,ro_y,test_size=0.2,random_state=42)
```

```python
lr2 = LogisticRegression()
lr2.fit(x_train,y_train)
```

```python
lr2.score(x_test,y_test)*100
```

```python
lr2.predict([[9.75, 67.53]])
```
