## NAIVEBAYES

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from mlxtend.plotting import plot_decision_regions
```

```python
ds = pd.read_excel(r"C:\Users\student\Downloads\SMLP Practicals\Excel data\placement_logistic_data.xlsx")
```

```python
ds.head(10)
```

```python
sns.kdeplot(data = ds["CGPA"])
```

```python
sns.scatterplot(x="CGPA",y="Score",data=ds,hue="Placed")
plt.legend(loc=1)
```

```python
x=ds.iloc[:,:-1]
y = ds["Placed"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.naive_bayes import GaussianNB, MultinomialNB, BernoulliNB
```

```python
gnb = GaussianNB()
gnb.fit(x_train,y_train)
```

```python
gnb.score(x_train,y_train)*100, gnb.score(x_test,y_test)*100
```

```python
gnb.predict([[9.75, 67.53]])
```

```python
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=gnb)
```

```python
mnb = MultinomialNB()
mnb.fit(x_train,y_train)
```

```python
mnb.score(x_train,y_train)*100, mnb.score(x_test,y_test)*100
```

```python
mnb.predict([[9.75, 67.53]])
```

```python
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=mnb)
```

```python
bnb = BernoulliNB()
bnb.fit(x_train,y_train)
```

```python
bnb.score(x_train,y_train)*100, bnb.score(x_test,y_test)*100
```

```python
bnb.predict([[9.75, 67.53]])
```

```python
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=bnb)
```
