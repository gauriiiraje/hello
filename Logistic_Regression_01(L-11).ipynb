## Cost Function

**Cost Function is help to make best fit line**
1. Cost function is an important parameter that determines how well a machine learning model performs for a given dataset.
2. Cost function is a measure of how wrong the model is in estimating the relationship between X(input) and Y(output) parameter.

   Types of COST Function:
   1. Regression Cost Function
   2. Classification cost function

**Binary Classification Cost Function :**
Classification models are used to make predictions of categorical variables,such as predictions for 0 or 1,cat or dog, etc. 

**Multi-class classifiation cost function:**
Multi-class classifiation cost functionis used in the classification problems for which instances are allocated to one of more than two class

## Logistic Regression

1. Logistic regression is one of the most popular Machine Learning algorithms, which comes under the Supervised Learning technique.

2. It is used for predicting the categorical dependent variable using a given set of independent variables.

3. Therefore, the outcome must be a categorical or discrete value.It can be either Yes or No, 0 or 1, true or false,etc. but instead of giving the exact value as 0and1, It gives the probabilistc values which lie between 0 and 1.

On the basis of the categories, Logistic Regression can be into three Types: 
- **Binomial**
-  
**Multinomial**
-
**Ordinal**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_excel(r"C:\Users\dell\Downloads\logistic_regression_data.xlsx")
```

```python
dataset.head()
```

```python
dataset.drop(columns=["Gender","CreditScore","Balance","EstimatedSalary","Tenure","IsActiveMember"],inplace=True)
```

```python
dataset.head()
```

```python
plt.figure(figsize=(6,3))
sns.scatterplot(x='Age',y='Purchased',data=dataset)
plt.show()
```

```python
x = dataset[["Age"]]
y = dataset["Purch"
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)*100
```

```python
lr.predict([[62]])
```

```python
plt.figure(figsize=(6,3))
sns.scatterplot(x='Age',y='Purchased',data=dataset)
sns.scatterplot(x = "Age",y = lr.predict(x),data=dataset,color="red")
plt.show()
```
