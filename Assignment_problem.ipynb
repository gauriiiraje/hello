```python
import numpy as np
from scipy.optimize import linear_sum_assignment
```

```python
cost = np.array([
    [9,2,7,8],
    [6,4,3,7],
    [5,8,1,8],
    [7,6,9,4]
])
```

```python
row, col = linear_sum_assignment(cost)
print("Assignment")
for r, c in zip(row, col):
    print(f"worker-{r+1} --> task-{c+1}")

print("Minimum Cost: ",cost[row,col].sum())
```
