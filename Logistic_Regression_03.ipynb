## Logistic Regression with Polynomial

```python
import pandas as pd
import matplotlib.pyplot as plt
import sklearn as sk
import seaborn as sns
```

```python
ds = pd.read_excel(r"C:\Users\student\Downloads\SMLP Practicals\Excel data\placement_logistic_data.xlsx")
```

```python
ds.head()
```

```python
sns.scatterplot(x="CGPA",y="Score",data=ds,hue="Placed")
plt.legend(loc=1)
```

```python
x=ds.iloc[:,:-1]
y = ds["Placed"]
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
pf = PolynomialFeatures(degree=3)
pf.fit(x)
x=pf.transform(x)
```

```python
#convert data from aray to dataframe
x=pd.DataFrame(x)
x
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=82)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)*100
```
