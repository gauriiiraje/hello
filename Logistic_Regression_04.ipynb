# Logistic Regression with Polynomial Data

```python
import pandas as pd
import matplotlib.pyplot as plt
import sklearn as sk
import seaborn as sns
```

```python
ds = pd.read_excel(r"C:\Users\student\Downloads\SMLP Practicals\Excel data\logistic_regression_multiclass_100.xlsx")
ds.head()
```

```python
ds["Placement_Level"].unique()
```

```python
sns.pairplot(data=ds,hue="Placement_Level")
plt.show()
```

```python
x=ds.iloc[:,:-1]
y = ds["Placement_Level"]
```

```python
from sklearn.model_selection import train_test_split 
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=82)
from sklearn.linear_model import LogisticRegression
```

## OVR

```python
#OVR converted data into binomial
lr = LogisticRegression(multi_class="ovr")
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)*100
```

```python
lr.predict([[6.69,41.7,4.2]])
```

## Multinomial

```python
#Multinomial converted data into binomial
lr = LogisticRegression(multi_class="multinomial")
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)*100
```

```python
lr.predict([[6.69,41.7,4.2]])
```
