```python
dataset["total_bill"]
```

```python
dataset["total_bill"].mean()
```

```python
mn = np.mean(dataset["total_bill"])
mn
```

```python
sns.histplot(x="total_bill",data=dataset,bins=[i for i in range(0,60,5)])
plt.show()
```

```python
sns.histplot(x="total_bill",data=dataset,bins=[i for i in range(0,60,5)])
plt.plot([mn for i in range(0,80)],[i for i in range(0,80)],c="red")
plt.show()
```

```python
dataset["total_bill"].median()
```

```python
md = np.median(dataset["total_bill"])
md
```

```python
sns.histplot(x="total_bill",data=dataset,bins=[i for i in range(0,60,5)])
plt.plot([md for i in range(0,80)],[i for i in range(0,80)],c="green")
plt.show()
```

```python
dataset["total_bill"].mode()
```

```python
type(dataset["total_bill"].mode())
```

```python
dataset["total_bill"].mode()[0]
```

```python
dataset["total_bill"].value_counts()
```

```python
mo=dataset["total_bill"].mode()[0]
```

```python
sns.histplot(x="total_bill",data=dataset,bins=[i for i in range(0,60,5)])
plt.plot([mo for i in range(0,80)],[i for i in range(0,80)],c="pink")
plt.show()
```

```python
sns.histplot(x="total_bill",data=dataset,bins=[i for i in range(0,60,5)])
plt.plot([mn for i in range(0,80)],[i for i in range(0,80)],c="red",label="mean")
plt.plot([md for i in range(0,80)],[i for i in range(0,80)],c="green",label="median")
plt.plot([mo for i in range(0,80)],[i for i in range(0,80)],c="purple",label="mode")
plt.legend()
plt.show()
```
